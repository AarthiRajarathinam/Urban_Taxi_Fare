# 🎯 Comprehensive Exploratory Data Analysis (EDA)
## Taxi Fare Prediction Dataset

This notebook performs detailed exploratory data analysis on the taxi fare dataset, including:
- **Univariate Analysis**: Distribution analysis of individual features
- **Bivariate Analysis**: Relationships between variables
- **Statistical Testing**: Correlation and pattern discovery
- **Data Quality Checks**: Validation and outlier detection
- **Visualizations**: Distribution plots, correlation matrices, and insights

### Dataset: NYC Taxi Fare Prediction
- **Target Variable**: Fare Amount (USD)
- **Features**: Trip characteristics, time information, and derived metrics
- **Observations**: 156,400+ valid taxi trips

```python
# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set visualization style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 6)
plt.rcParams['font.size'] = 10

print("✓ All libraries imported successfully!")
```

## Section 1: Load and Explore Dataset

```python
# Load the engineered dataset
df = pd.read_csv(r'C:\Aarthi\Guvi\MINI_PROJECT_3\taxi_fare_engineered.csv')

print("=" * 80)
print("DATASET OVERVIEW")
print("=" * 80)
print(f"\nDataset Shape: {df.shape}")
print(f"Total Rows: {len(df):,}")
print(f"Total Columns: {len(df.columns)}")

print("\n" + "=" * 80)
print("DATA TYPES")
print("=" * 80)
print(df.dtypes)

print("\n" + "=" * 80)
print("MISSING VALUES")
print("=" * 80)
missing = df.isnull().sum()
if missing.sum() > 0:
    missing_pct = (missing / len(df) * 100).round(2)
    missing_df = pd.DataFrame({'Missing Count': missing[missing > 0], 'Percentage': missing_pct[missing > 0]})
    print(missing_df)
else:
    print("✓ No missing values found!")

print("\n" + "=" * 80)
print("STATISTICAL SUMMARY")
print("=" * 80)
print(df.describe())
```

## Section 2: Data Preparation and Validation

```python
# Filter to valid trips for analysis
df_valid = df[(df['is_valid_trip'] == 1) & 
              (df['trip_duration_minutes'] > 0) & 
              (df['speed_mph'] < 100)].copy()

print("=" * 80)
print("DATA VALIDATION")
print("=" * 80)
print(f"\nOriginal dataset: {len(df):,} rows")
print(f"Valid trips for analysis: {len(df_valid):,} rows ({len(df_valid)/len(df)*100:.2f}%)")
print(f"Rows removed: {len(df) - len(df_valid):,}")

# Calculate derived metrics
df_valid['fare_per_mile'] = np.where(
    df_valid['trip_distance_miles'] > 0,
    df_valid['fare_amount'] / df_valid['trip_distance_miles'],
    0
)

df_valid['fare_per_minute'] = np.where(
    df_valid['trip_duration_minutes'] > 0,
    df_valid['fare_amount'] / df_valid['trip_duration_minutes'],
    0
)

print("\n✓ Derived metrics calculated successfully!")
```

## Section 3: Univariate Analysis - Distribution of Key Variables

```python
# Univariate Analysis
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle('Univariate Analysis - Key Variables Distribution', fontsize=16, fontweight='bold')

# Fare Amount
axes[0, 0].hist(df_valid['fare_amount'], bins=50, color='steelblue', edgecolor='black')
axes[0, 0].set_title('Fare Amount Distribution')
axes[0, 0].set_xlabel('Fare ($)')
axes[0, 0].set_ylabel('Frequency')

# Trip Duration
axes[0, 1].hist(df_valid['trip_duration_minutes'], bins=50, color='coral', edgecolor='black')
axes[0, 1].set_title('Trip Duration Distribution')
axes[0, 1].set_xlabel('Duration (minutes)')
axes[0, 1].set_ylabel('Frequency')

# Trip Distance
axes[0, 2].hist(df_valid['trip_distance_miles'], bins=50, color='lightgreen', edgecolor='black')
axes[0, 2].set_title('Trip Distance Distribution')
axes[0, 2].set_xlabel('Distance (miles)')
axes[0, 2].set_ylabel('Frequency')

# Fare per Mile
axes[1, 0].hist(df_valid['fare_per_mile'], bins=50, color='gold', edgecolor='black')
axes[1, 0].set_title('Fare per Mile Distribution')
axes[1, 0].set_xlabel('Fare per Mile ($/mi)')
axes[1, 0].set_ylabel('Frequency')

# Pickup Hour
axes[1, 1].hist(df_valid['pickup_hour'], bins=24, color='purple', edgecolor='black')
axes[1, 1].set_title('Pickup Hour Distribution')
axes[1, 1].set_xlabel('Hour of Day')
axes[1, 1].set_ylabel('Frequency')

# Passenger Count
axes[1, 2].hist(df_valid['passenger_count'], bins=7, color='teal', edgecolor='black')
axes[1, 2].set_title('Passenger Count Distribution')
axes[1, 2].set_xlabel('Passenger Count')
axes[1, 2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

print("\n✓ Univariate analysis visualizations completed!")
```

## Section 4: Bivariate Analysis - Relationships Between Variables

```python
# Bivariate Analysis - Scatter plots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Bivariate Analysis - Relationship with Fare Amount', fontsize=16, fontweight='bold')

# Fare vs Distance
axes[0, 0].scatter(df_valid['trip_distance_miles'], df_valid['fare_amount'], alpha=0.5, s=20)
axes[0, 0].set_title('Fare vs Trip Distance')
axes[0, 0].set_xlabel('Distance (miles)')
axes[0, 0].set_ylabel('Fare ($)')

# Fare vs Duration
axes[0, 1].scatter(df_valid['trip_duration_minutes'], df_valid['fare_amount'], alpha=0.5, s=20, color='orange')
axes[0, 1].set_title('Fare vs Trip Duration')
axes[0, 1].set_xlabel('Duration (minutes)')
axes[0, 1].set_ylabel('Fare ($)')

# Fare vs Pickup Hour
axes[1, 0].scatter(df_valid['pickup_hour'], df_valid['fare_amount'], alpha=0.5, s=20, color='green')
axes[1, 0].set_title('Fare vs Pickup Hour')
axes[1, 0].set_xlabel('Hour of Day')
axes[1, 0].set_ylabel('Fare ($)')

# Fare vs Passenger Count
axes[1, 1].scatter(df_valid['passenger_count'], df_valid['fare_amount'], alpha=0.5, s=20, color='red')
axes[1, 1].set_title('Fare vs Passenger Count')
axes[1, 1].set_xlabel('Passenger Count')
axes[1, 1].set_ylabel('Fare ($)')

plt.tight_layout()
plt.show()

print("\n✓ Bivariate analysis visualizations completed!")
```

## Section 5: Correlation Analysis

```python
# Correlation Analysis
numeric_cols = df_valid.select_dtypes(include=[np.number]).columns.tolist()
correlation_matrix = df_valid[numeric_cols].corr()

# Display correlation with fare_amount
print("=" * 80)
print("CORRELATION WITH FARE AMOUNT (Target Variable)")
print("=" * 80)
fare_corr = correlation_matrix['fare_amount'].sort_values(ascending=False)
print(fare_corr)

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f', square=True, cbar_kws={'label': 'Correlation'})
plt.title('Correlation Matrix of Numeric Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n✓ Correlation analysis completed!")
```

## Section 6: Outlier Detection

```python
# Outlier Detection using IQR method
print("=" * 80)
print("OUTLIER DETECTION")
print("=" * 80)

def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (data[column] < lower_bound) | (data[column] > upper_bound)

# Detect outliers in key columns
outlier_cols = ['fare_amount', 'trip_duration_minutes', 'trip_distance_miles']
outlier_summary = {}

for col in outlier_cols:
    outliers = detect_outliers_iqr(df_valid, col)
    outlier_summary[col] = {
        'outlier_count': outliers.sum(),
        'percentage': (outliers.sum() / len(df_valid) * 100)
    }
    print(f"\n{col}:")
    print(f"  Outliers found: {outliers.sum()} ({outlier_summary[col]['percentage']:.2f}%)")
    print(f"  Value range: [{df_valid[col].min():.2f}, {df_valid[col].max():.2f}]")

print("\n✓ Outlier detection completed!")
```
